In this notebook we demonstrate how the example dataset "branchoid" was made.

```python
import pyclesperanto_prototype as cle
import napari
import numpy as np
import napari_process_points_and_surfaces as nppas
import vedo
```

```python
size = 100

image = cle.create([size, size, size])
```

```python
cle.set(image, 0)
```

```python
cle.draw_sphere(image, size*0.5, size*0.5, size*0.5, radius_x=size*0.25, radius_y=size*0.25, radius_z=size*0.25)
```

```python
for i in range(0, 10):
    
    radius = size * (0.25 - i*0.01)
    
    cle.draw_sphere(image, 
                    size*0.5 + i * size * 0.02, 
                    size*0.5, 
                    size*0.5, 
                    radius_x=radius, 
                    radius_y=radius, 
                    radius_z=radius)

image
```

```python
for i in range(0, 10):
    
    radius = size * (0.15 - i*0.01)
    
    cle.draw_sphere(image, 
                    size*0.5, 
                    size*0.65 + i * size * 0.02, 
                    size*0.5, 
                    radius_x=radius, 
                    radius_y=radius, 
                    radius_z=radius)

image
```

```python
for i in range(0, 10):
    
    radius = size * (0.12 - i*0.01)
    
    cle.draw_sphere(image, 
                    size*0.35 - i * size * 0.02, 
                    size*0.65 + i * size * 0.02, 
                    size*0.5, 
                    radius_x=radius, 
                    radius_y=radius, 
                    radius_z=radius)

image
```

```python
for i in range(0, 10):
    
    radius = size * (0.05)
    
    cle.draw_sphere(image, 
                    size*0.35 - i * size * 0.02, 
                    size*0.5, 
                    size*0.5, 
                    radius_x=radius, 
                    radius_y=radius, 
                    radius_z=radius)

image
```

```python
for i in range(0, 20):
    
    if i < 10:
        radius = size * (0.05)
    else:
        radius = size * (0.05 + (i-10)*0.01)
    
    cle.draw_sphere(image, 
                    size*0.35 - i * size * 0.01, 
                    size*0.35 - i * size * 0.01, 
                    size*0.5, 
                    radius_x=radius, 
                    radius_y=radius, 
                    radius_z=radius)

image
```

```python
viewer = napari.Viewer(ndisplay=3)
viewer.add_labels(image.astype(int))
```

```python
surface = nppas.label_to_surface(image)

viewer.add_surface(surface)

napari.utils.nbscreenshot(viewer)
```
