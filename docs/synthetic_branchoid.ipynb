```python
import pyclesperanto_prototype as cle
import numpy as np
```

```python
image = cle.create((100, 100, 100))
```

```python
cle.set(image, 0)

cle.draw_sphere(image, 
                x=50,
                y=50,
                z=50,
                radius_x=20,
                radius_y=20,
                radius_z=20
               )

cle.draw_sphere(image, 
                x=20,
                y=50,
                z=50,
                radius_x=5,
                radius_y=5,
                radius_z=5
               )

cle.draw_sphere(image, 
                x=50,
                y=25,
                z=50,
                radius_x=5,
                radius_y=5,
                radius_z=5
               )

cle.draw_sphere(image, 
                x=70,
                y=50,
                z=50,
                radius_x=5,
                radius_y=5,
                radius_z=5
               )

cle.draw_sphere(image, 
                x=50,
                y=85,
                z=50,
                radius_x=5,
                radius_y=5,
                radius_z=5
               )
```

```python
sigma = 5
binary = cle.gaussian_blur(image, 
                  sigma_x=sigma,
                  sigma_y=sigma,
                  sigma_z=sigma,
                 ) > 0.1
binary
```

```python
import napari

viewer = napari.Viewer(ndisplay=3)

viewer.add_labels(binary)

napari.utils.nbscreenshot(viewer)
```
